```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameters for the normal distribution
mean = 50
std_dev = 5
num_points = 1000

# Generate random data points from the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=num_points)

# Plot the histogram of the generated data
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=False, bins=30, color='skyblue', stat='density', label='Histogram')

# Overlay the Probability Density Function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r-', lw=2, label='PDF (Normal Distribution)')

# Label the plot
plt.title('Normal Distribution with Mean = 50, Standard Deviation = 5')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 14
sigma = 2.5
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')

# Shading areas
# I: Between 14 and 2.5
x1 = np.linspace(2.5, 14, 100)
y1 = norm.pdf(x1, mu, sigma)
plt.fill_between(x1, y1, color='green', alpha=0.5, label='14 > x > 2.5')

# II: Below 12
x2 = np.linspace(mu - 4*sigma, 12, 100)
y2 = norm.pdf(x2, mu, sigma)
plt.fill_between(x2, y2, color='red', alpha=0.5, label='x < 12')

# III: Above 19
x3 = np.linspace(19, mu + 4*sigma, 100)
y3 = norm.pdf(x3, mu, sigma)
plt.fill_between(x3, y3, color='orange', alpha=0.5, label='x > 19')

# Labels and legend
plt.title('Normal Distribution Visualization')
plt.xlabel('Scores')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
