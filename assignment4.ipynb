```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('titanic.csv')

# Plot histogram
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Age among Passengers')
plt.show()
```

```python
sns.lmplot(data=data, x='Age', y='Fare', hue='Survived', aspect=2)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatterplot with Trendline of Fare vs. Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Make sure seaborn is imported


# Load the dataset
data = pd.read_csv('titanic.csv')

# Handle missing values
data = data.dropna(subset=['Age', 'Fare'])

# Create scatterplot, change df to data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data)
plt.title('Scatterplot of Fare and Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survival Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
# Replace 'df' with 'data' which contains the titanic dataset
sns.countplot(x='Class', hue='Survived', data=data)
plt.title('Survival Rates by Pclass')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Survival Status')
plt.show()
```
