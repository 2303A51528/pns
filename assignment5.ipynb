```python
import numpy as np

# Number of households
N = 50

# Randomly generate the number of vehicles per household (0 to 5)
np.random.seed(42)  # For reproducibility
vehicles_owned = np.random.randint(0, 6, N)

print("Number of vehicles owned per household:", vehicles_owned)
```

```python
import matplotlib.pyplot as plt

# Compute the PMF
values, counts = np.unique(vehicles_owned, return_counts=True)
pmf = counts / N

# Compute the CDF
cdf = np.cumsum(pmf)

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.bar(values, pmf, width=0.5, color='blue', alpha=0.7, label='PMF')
plt.xlabel('Number of Vehicles')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF)')
plt.legend()
plt.show()

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(values, cdf, marker='o', linestyle='-', color='red', label='CDF')
plt.xlabel('Number of Vehicles')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.legend()
plt.show()
```

```python
import scipy.stats as stats

# Number of gadgets inspected
N = 100  # You can change this value as needed

# Probability of a gadget being defective
p = 0.05

# i) Probability that exactly 10 gadgets are defective
k = 10
prob_10_defective = stats.binom.pmf(k, N, p)
print(f"Probability that exactly 10 gadgets are defective: {prob_10_defective:.4f}")

# ii) Probability that at least 15 gadgets are defective
k_min = 15
prob_at_least_15_defective = 1 - stats.binom.cdf(k_min - 1, N, p)
print(f"Probability that at least 15 gadgets are defective: {prob_at_least_15_defective:.4f}")

# iii) Average number of defective gadgets
expected_defective = N * p
print(f"Average number of defective gadgets: {expected_defective:.2f}")
```
